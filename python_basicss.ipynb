```python
#create a student class that takes name and marks of three subjects as an argument in conductor. then create a method to find average marks of the student and print it
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def find_average(self):
        total = sum(self.marks)
        average = total / len(self.marks)
        print(f"Average marks for {self.name}: {average}")
student1 = Student("Preet", [85, 90, 95])
student1.find_average()
del student1.name
```

```python
#inheritance is a feature of object-oriented programming that allows a new class (called a child class or subclass) to inherit properties and behaviors (attributes and methods) from an existing class (called a parent class or superclass). The child class can also have its own unique attributes and methods, in addition to those inherited from the parent class.
class car : #this is the parent class of maruti and tuype class known as base class
    color="black"
    
    def __init__(self,engine):
        self.engine=engine #this is the constructor of thr car class
        
    
    @staticmethod #static method is a decorator used to define a method inside a class that does not require access to the class (via cls) or its instances (via self)
    def start():
        print("car is starting")
         
    @staticmethod # it cannot access or modify class state and they are generally used to perform utility functions when there is no attribute needed
    def stop():
        print("car is stopping") #no attribute

class maruti(car): #maruti is the child class of car class known as derived class 
    def __init__(self,name,model,engine):
        self.name=name
        self.model=model #this is a attribute
        super().__init__(engine) #super()is used to call the constructor of the parent class
        print("creating new car")
        
#till this point this inheritance is a single level inheritance
class types(maruti):#types is the child class of maruti class and grandchild class of car class known as derived class
    def __init__(self,type):
        self.type=type
        
#this is a multilevel inheritance we can add more levels as per choice and it will still be known as multilevel ingheritance       
maruti1= maruti("Maruti", "Swift","2000cc")
maruti2= maruti("Maruti", "800","1500cc")
types1= types("petrol")
print (maruti1.name, maruti1.model, maruti1.engine)
print (maruti2.name, maruti2.model, maruti2.engine)
print(maruti1.start())
print(maruti1.stop())
print(maruti2.color)
print(types1.type)
```

```python
#property decorator used to define method/function as a property and use it as attribute
# gives us getter(to read) ,setter(to write) and deleter(to delete) meathods
class rectangle:   
    def __init__(self,width,height):
        self._width=width
        self._height=height
    @property   
    def width(self):
        return {f" width of rectangle is {self._width}cm"}
    @property
    def height(self):
        return {f" height of rectangle is {self._height}cm"}

rect1= rectangle(10,5)
print(rect1.width)
    
```

```python
#property decorator used to define method/function as a property and use it as attribute
# gives us getter(to read) ,setter(to write) and deleter(to delete) meathods
class Student:
    def __init__(self ,sub1,sub2,sub3): 
        self.sub1=sub1
        self.sub2=sub2
        self.sub3=sub3
        
    @property #property decorator is used to define a method as a property which can be accessed like an attribute without calling it as a method
    def percentage(self):
        return str((self.sub1 + self.sub2 + self.sub3)/3) + "%"
student1= Student(85,90,95)
print(student1.percentage)      
student1.sub1=80
print(student1.percentage)
```

```python
#Polymorphism :opertator overloading
#when the same operator is allowed to have different meanings according to the context.
#operator and dunder functions
#a+b addition         a.__add__(b)
#a-b subtraction      a.__sub__(b)
#a*b multiplication   a.__mul__(b)
#a/b division         a.__truediv__(b)
#a%b modulus          a.__mod__(b)
print(1+2) #this is addition
print("Hello"+" "+"World") #this is string concatenation
print ([1,2]+[3,4]) #this is list concatenation
#in here operator has diff meanings according to context
class complex:
    def __init__(self,real,imag):
        self.real=real
        self.imag=imag
    def shownum(self):
        print(self.real,"i","+",self.imag,"j")
    def __add__(self,other): #this operator allows ua to add 2 num without function .add and can be used
        return complex(self.real + other.real, self.imag + other.imag)
c1= complex(2,3)
c2= complex(4,5)
c1.shownum()
c2.shownum()
c3= c1+c2 #this will call the __add__ method and add the real and imaginary parts of c1 and c2
c3.shownum()
```

```python
#abstract base class ABC 
#abstact base class  is a class that cannot be instantiated.
#ABC cannot be used until it is a parent class
#ABC's child class must implement all the abstract methods of the ABC class
from abc import ABC, abstractmethod #abc is a module in python that provides the infrastructure for defining abc
class veichle(ABC): #veichle is the abstract base class
    @abstractmethod #abstractmethod is a decorator that is used to define abstract methods
    def start(self):
        pass #pass is a null statement that does nothing
    @abstractmethod
    def stop(self):
        pass
class car(veichle): #car is the child class of veichle class and it must implement all the abstract methods of veichle class
    def start(self):
        print("car is starting")
    def stop(self):
        print("car is stopping") 
#car has to inherit form veichle class and implement all the functions/abstract methods of veichle class otherwise it will give an error
#child class of ABC is known as concrete class.
class bike(veichle):
    def start(self):
        print("bike is starting")
    def stop(self):
        print("bike is stopping") #if we forget to add any of the abstract method in child class the erroe will specify which method is not implemented in child class
#this is the main advantage of using ABC because it ensures that all the necessary methods are implemented in the child class and it provides a clear structure for the code
car1= car()
bike1= bike()
car1.start()
car1.stop()
bike1.start()
bike1.stop()
```

```python
#encapsulation is the process of hiding the internal details of an object and only exposing a public interface for interacting with the object. It is one of the fundamental principles of object-oriented programming (OOP) and helps to protect the integrity of the data and prevent unauthorized access.

class payment:
    
    def __init__(self, amount):
        self.__amount = amount +amount*0.12 #this is a private variable because it is defined with __ before the variable name and it cannot be accessed outside the class
    
    def get_amount(self):#this is a getter method that allows us to access the value of the private variable __amount from outside the class
        return self.__amount
    
    def set_amount(self,discount):#this is a setter method that allows us to set the value of the private variable __amount from outside the class
        self.__amount = self.__calculate_discount(discount)
    
    def __calculate_discount(self,discount):#this is a private method that allows us to calculate the discount on the amount and no one can use this method  out of the class.
        return self.__amount - self.__amount*(discount/100)
    
payment1= payment(1000)
print(payment1.get_amount()) #this will print the amount with gst
payment1.set_amount(10) #this will set the amount with 10% discoun
#this process of hiding the internal details of an objects lets us to protect the integrity of the data and prevent unauthorized access.
#this is known as encapsulation because it encapsulates the data and the methods that operate on the data within a single unit (the class) and restricts access to the internal details of the object. 
```
