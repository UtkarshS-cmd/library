```python
from ast import While
import json
from pathlib import Path

contact_file = Path("contacts.json") # This line creates a Path object representing the file "contacts.json". This file will be used to store the contact information in JSON format. The Path class from the pathlib module provides an easy way to work with file paths and perform file operations.

def load_contacts():
    """Load contacts from JSON file"""
    if contact_file.exists():
        with contact_file.open("r", encoding="utf-8") as f:# r here means read mode, and encoding="utf-8" ensures that the file is read using UTF-8 encoding, which can handle a wide range of characters.
            return json.load(f) #jason.load() reads the JSON data and converts it into a Python list of dictionaries (contacts).
    return []


def save_contacts(contacts):#contancts is a parent function that has list of dictionaries of contacts.
    with contact_file.open("w", encoding="utf-8") as f: # w here represents write mode.
        json.dump(contacts, f, indent=4) #json.dump() takes the list of contacts and writes it to the file in JSON format. 

def create_contact():
    while True:
        print("\n--- Add New Contact --- (type 'menu' to return to menu)")
        name = input("Enter name: ").strip()
        if name.lower() == 'menu':
            print("Returning to menu.")
            return
        if not name:
            print("Name cannot be empty!")
            continue
        while True:
            phone = input("Enter phone: ").strip() #strip() is used to remove any leading or trailing whitespace from the input. This ensures that the phone number is stored without any extra spaces, which can help maintain data consistency and prevent issues when searching or displaying contacts.
            if phone.lower() == 'menu':
               print("Returning to menu.")
               return
            if not phone:
                print("Phone cannot be empty!")
                continue
            if not phone.isdigit(): #isdigit() is a string method that checks if all characters in the string are digits. If the phone number contains any non-digit characters (like letters or special characters), this check will fail, and the program will prompt the user to enter a valid phone number.
               print("Phone must be a number!")
               continue
            if len(phone) != 10:
                print("Phone must be 10 digits!")
                continue
        break
    
        while True:
            email = input("Enter email: ").strip()
            if email.lower() == 'menu':
                print("Returning to menu.")
                return
            if not email:
                print("Email cannot be empty!")
                continue
            if "@" not in email or "." not in email: # This check is a simple way to validate that the email address contains the basic components of a valid email (the "@" symbol and a period). If either of these components is missing, the program will prompt the user to enter a valid email address.
                print("Invalid email format!")
                continue
        break

    contacts = load_contacts()
    # Check if contact already exists
    for c in contacts:
        if c["name"].lower() == name.lower(): #lower changes the name already in contact list and thr name entered now and compares both if it is same it does not save the contact.
            print(f"Contact '{name}' already exists!") #f means string fromatting it hepls to add the string and name variable in the same line.
            return

    contacts.append({ #append makes a new dictionary
        "name": name,
        "phone": phone,
        "email": email,
    })
    save_contacts(contacts)
    print(f" Contact '{name}' added successfully!") 


def read_contacts():
    contacts = load_contacts()
    if not contacts:#search if the list of contacts is empty or not if it is empty it will print no contacts found.
        
        print("\n No contacts found.")
        return
    
    print(f"\n--- All Contacts ({len(contacts)}) ---") #prints number of contacts in the list.
    for i, contact in enumerate(contacts, 1): #enumerate is used to loop through the list of contacts and it provides both the index (i) and the contact itself.
        print(f"{i}. Name: {contact['name']}, Phone: {contact['phone']}, Email: {contact['email']}")


def search_contact():
    print("\n--- Search Contact ---")
    name = input("Enter name to search: ").strip()
    if not name:
        print(" Name cannot be empty!")
        return
    
    contacts = load_contacts()
    found = [c for c in contacts if name.lower() in c["name"].lower()]
    
    if not found:
        print(f" No contacts found matching '{name}'")
        return
    
    print(f" Found {len(found)} match:")
    for contact in found:
        print(f"   Name: {contact['name']}, Phone: {contact['phone']}, Email: {contact['email']}")


def update_contact():
    print("\n--- Update Contact ---")
    name = input("Enter name of contact to update: ").strip()
    if not name:
        print(" Name cannot be empty!")
        return
    
    contacts = load_contacts()
    for contact in contacts:
        if contact["name"].lower() == name.lower():
            print(f"Updating '{contact['name']}'...")
            new_phone = input(f"Enter new phone (current: {contact['phone']}): ").strip()
            new_email = input(f"Enter new email (current: {contact['email']}): ").strip()
            
            if new_phone:
                contact["phone"] = new_phone
            if new_email:
                contact["email"] = new_email
            
            save_contacts(contacts)
            print(f" Contact '{contact['name']}' updated successfully!")
            return
    
    print(f" Contact '{name}' not found.")


def delete_contact():
    """Delete a contact"""
    print("\n--- Delete Contact ---")
    name = input("Enter name of contact to delete: ").strip()
    if not name:
        print(" Name cannot be empty!")
        return
    
    contacts = load_contacts()
    for i, contact in enumerate(contacts):
        if contact["name"].lower() == name.lower():
            confirm = input(f"Delete '{contact['name']}'? (y/n): ").strip().lower()
            if confirm == "y":
                contacts.pop(i)# deletes and shows /pop the contact selected
                save_contacts(contacts)
                print(f"Contact '{contact['name']}' deleted successfully!")
            else:
                print(" Deletion cancelled.")
            return
    
    print(f"Contact '{name}' not found.")


def main():
    """Main menu loop"""
    print("=" * 40)
    print("   Contact Manager App")
    print("=" * 40)
    
    while True:
        print("\n--- Menu ---")
        print("1. Add Contact")
        print("2. View All Contacts")
        print("3. Search Contact")
        print("4. Update Contact")
        print("5. Delete Contact")
        print("6. Exit")
        
        choice = input("\nChoose an option (1-6): ").strip()
        
        if choice == "1":
            create_contact()
        elif choice == "2":
            read_contacts()
        elif choice == "3":
            search_contact()
        elif choice == "4":
            update_contact()
        elif choice == "5":
            delete_contact()
        elif choice == "6":
            print("\n Thank you for using Contact Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1-6.")


if __name__ == "__main__":
    main()
```
